We want to predict a binary variable - whether the salary was low or high. Compute the median salary and create a new binary variable that is true when the salary is high (above the median).
We could also perform Linear Regression (or any regression) to predict the salary value here. Instead, we are going to convert this into a binary classification problem, by predicting two classes, HIGH vs LOW salary.

While performing regression may be better, performing classification may help remove some of the noise of the extreme salaries. We don't have to choose the median as the splitting point - we could also split on the 75th percentile or any other reasonable breaking point.

In fact, the ideal scenario may be to predict many levels of salaries.

```python
# dummify only city
df_city_dum = pd.get_dummies(df['City'])
df_city_dum.head()
```

```python
# Define tatget and predicters only using the location as a feature
X = df_city_dum
y = df['Median']
```

```python
# Standardlize the scaling
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
X = scaler.fit_transform(X)
```

```python
# Classfier Model1 - Decision Tree
rfc = RandomForestClassifier(n_estimators=100, oob_score=True)
rfc.fit(X, y)
cv = cross_val_score(rfc, X, y, cv=5)

print('Cross-validated scores:', cv)
print('Average score:', cv.mean())
print('Standard deviation of score:', cv.std())
print (rfc.oob_score_)
```

```python
# Classfier Model2 - K Nearest Neighbors

knn = KNeighborsClassifier(n_neighbors=5)
knn.fit(X, y)
cv = cross_val_score(knn, X, y, cv=5)

print('Cross-validated scores:', cv)
print('Average score:', cv.mean())
print('Standard deviation of score:', cv.std())
# print (knn.oob_score_)
```

```python
# Classfier Model3 - K Means
from sklearn.cluster import KMeans
k_means = KMeans(n_clusters=3, random_state=1)

k_means.fit(X, y)
cv = cross_val_score(k_means, X, y, cv=5)

print('Cross-validated scores:', cv)
print('Average score:', cv.mean())
print('Standard deviation of score:', cv.std())
```
